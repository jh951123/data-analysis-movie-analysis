```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import csv
```

```python
user_data = pd.read_csv('./data/naver_user.csv')
user_data.tail(5)
```

```python
from pandas import Series, DataFrame
import numpy as np
import re

rating = DataFrame({'userId':[], 'rating':[], 'movieId':[]})
rating
```

```python
import time
import random

URL = "https://movie.naver.com/movie/point/af/list.nhn?st=nickname&sword="
#URL = "https://movie.naver.com/movie/point/af/list.nhn?st=nickname&sword=15774287&target=after"

for j in range(100):
    user_id = user_data.iloc[j, 0]
    req = requests.get(URL + str(user_id) + '&target=after&page=' + '1')
    html = req.text
    user_temp = user_data[user_data['reviewNo'] == user_id]
   
    print(user_temp)

    user = user_temp.iloc[0, 1]
    soup = BeautifulSoup(html, 'html.parser')

    spanedList = soup.select('.c_88')

    page_num = (int(spanedList[0].text)//10) + 1
        
    for i in range(page_num):
        req1 = requests.get(URL + str(user_id) + '&target=after&page=' + str(i+1))
        html1 = req1.text

        soup1 = BeautifulSoup(html1, 'html.parser')
        keywordList = soup1.select('.list_netizen tbody tr') # 불러올 태그에 대한 셀렉터 지정.

        for key in keywordList:
            temp = []
            temp.append(user)
            temp.append(int(key.select(".point")[0].text))
            href = key.select(".title > a")[0]['href']
            reviewNo = int(key.select(".ac.num")[0].text)
            
            if(reviewNo <= int(user_id)):
                movieId = re.findall("\d+", href)

                temp.append(int(movieId[0]))
        #            print("사람", key.select(".num > .author")[0].text) # 태그 안에 있는 텍스트만 전부 가져올 수 있다.
                print(temp)
                length = len(rating)
                rating.loc[length] = temp
        print("------------------------------------")
    time.sleep(random.randrange(3, 5))
```

```python
rating
```

```python
result = rating

result = result.drop_duplicates()
result
```

```python
user_num = result.userId.unique().shape[0]
movie_num = result.movieId.unique().shape[0]

print("유저의 수는 {}명 이고, \n영화의 수는 {}개 입니다.".format(user_num, movie_num))
```

```python
result.to_csv("./data/rating.csv")
```
